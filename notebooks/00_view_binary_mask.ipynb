Helper notebook to view binary masks (e.g. vessel annotations) in grayscale

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image


def show_binary_mask(path: str | Path) -> None:
    """
    Load a binary PNG mask, scale it to 0-255 grayscale for viewing,
    and display it.
    """

    # Load mask
    mask = np.array(Image.open(path))

    # Convert to binary if needed, then scale to grayscale
    grayscale = (mask > 0).astype(np.uint8) * 255

    # Display
    plt.figure(figsize=(6, 6))
    plt.imshow(grayscale, cmap="gray", vmin=0, vmax=255)
    plt.axis("off")
    plt.show()
```
